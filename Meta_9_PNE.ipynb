# Análise dos impactos da pandemia de Covid-19 no cumprimento da Meta 9 do Plano Nacional de Educação (PNE)

Quanto à Meta 9, seu objetivo é a elevação da taxa de alfabetização da população
de 15 anos ou mais para 93,5% até 2015 e, ao final da vigência do PNE, erradicar o
analfabetismo absoluto e reduzir em 50% a taxa de analfabetismo funcional.

Os dados em relação à faixa etária acima dos 40 anos indicam que esta Meta
revela grande desafio a ser superado. Em relação à taxa de analfabetismo na faixa
etária entre 15 e 19 anos, 0,7%, o que se observa na faixa entre os 50 e 59 anos é uma
taxa cinco vezes maior, e acima dos 60 anos, é dezoito vezes maior.

Fonte: IBGE - Pesquisa Nacional por Amostra de Domicílios Contínua Anual - 2º trimestre

Tabela 7113 - Taxa de analfabetismo das pessoas de 15 anos ou mais de idade, por sexo e grupo de idade"
"Variável - Taxa de analfabetismo das pessoas de 15 anos ou mais de idade (%)"

Taxa de analfabetismo é a razão entre o número de pessoas de determinada faixa etária que não sabem ler e escrever um recado ou bilhete simples no idioma que conhecem e o total de pessoas dessa mesma faixa etária.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('default')
df = pd.read_csv('/content/analfabetismo_idades_regioes.csv', sep=';', decimal=',')
df
```

# 2018

```python
df_brasil = df.loc[df['Brasil e Grande Região'] == "Brasil"]
df_brasil_2018 = df_brasil.loc[df['Ano'] == 2018]
# df_brasil_2018 = df_brasil_2018[['Grupo de idade', 'Taxa']]
# df_brasil_2018 = df_brasil_2018.T
df_brasil_2018
```

```python
df_brasil_2018_t = df_brasil_2018.loc[df_brasil_2018['Sexo'] == 'Total']
df_brasil_2018_t = df_brasil_2018_t[['Grupo de idade', 'Taxa']]
df_brasil_2018_t.rename(columns = {'Taxa':'Total'}, inplace = True)
df_brasil_2018_t = df_brasil_2018_t.T
df_brasil_2018_t
```

```python
df_brasil_2018_m = df_brasil_2018.loc[df_brasil_2018['Sexo'] == 'Mulheres']
df_brasil_2018_m = df_brasil_2018_m[['Grupo de idade', 'Taxa']]
df_brasil_2018_m.rename(columns = {'Taxa':'Mulheres'}, inplace = True)
df_brasil_2018_m = df_brasil_2018_m.T
df_brasil_2018_m = df_brasil_2018_m.iloc[1:]
df_brasil_2018_m
```

```python
df_brasil_2018_h = df_brasil_2018.loc[df_brasil_2018['Sexo'] == 'Homens']
df_brasil_2018_h = df_brasil_2018_h[['Grupo de idade', 'Taxa']]
df_brasil_2018_h.rename(columns = {'Taxa':'Homens'}, inplace = True)
df_brasil_2018_h = df_brasil_2018_h.T
df_brasil_2018_h = df_brasil_2018_h.iloc[1:]
df_brasil_2018_h
```

```python
df_brasil_2018_t.columns = ['0', '1', '2', '3', '4']
df_brasil_2018_h.columns = ['0', '1', '2', '3', '4']
df_brasil_2018_m.columns = ['0', '1', '2', '3', '4']
```

```python
frames = [df_brasil_2018_t, df_brasil_2018_h, df_brasil_2018_m]
df_brasil_2018_c = pd.concat(frames)
df_brasil_2018_c
```

```python
df_brasil_2018_c = df_brasil_2018_c.T
df_brasil_2018_c
```

```python
df_brasil_2018_c = df_brasil_2018_c.set_index('Grupo de idade')
```

```python
df_brasil_2018_c.plot(kind='bar')

plt.title("Taxa de analfabetismo no Brasil em 2018, por sexo e grupo de idade")
plt.xlabel("Grupo de Idade")
plt.ylabel("Taxa de Analfabetismo")

plt.xticks(rotation=30, horizontalalignment="center")

plt.show()
```

# Regiões

```python
df_sul = df.loc[df['Brasil e Grande Região'] == "Sul"]
df_sul = df_sul.loc[df['Sexo'] == 'Total']
df_sul = df_sul.loc[df['Grupo de idade'] == '15 anos ou mais']
# df_sul_2018 = df_sul.loc[df['Ano'] == 2018]
df_sul = df_sul[['Ano', 'Taxa']]
df_sul.rename(columns = {'Taxa':'Sul'}, inplace = True)
df_sul = df_sul.reset_index()
df_sul = df_sul[['Ano', 'Sul']]
df_sul
```

```python
df_norte = df.loc[df['Brasil e Grande Região'] == "Norte"]
df_norte = df_norte.loc[df['Sexo'] == 'Total']
df_norte = df_norte.loc[df['Grupo de idade'] == '15 anos ou mais']
df_norte = df_norte[['Taxa']]
df_norte.rename(columns = {'Taxa':'Norte'}, inplace = True)
df_norte = df_norte.reset_index()
df_norte = df_norte[['Norte']]
df_norte
```

```python
df_nordeste = df.loc[df['Brasil e Grande Região'] == "Nordeste"]
df_nordeste = df_nordeste.loc[df['Sexo'] == 'Total']
df_nordeste = df_nordeste.loc[df['Grupo de idade'] == '15 anos ou mais']
df_nordeste = df_nordeste[['Taxa']]
df_nordeste.rename(columns = {'Taxa':'Nordeste'}, inplace = True)
df_nordeste = df_nordeste.reset_index()
df_nordeste = df_nordeste[['Nordeste']]
df_nordeste
```

```python
df_sudeste = df.loc[df['Brasil e Grande Região'] == "Sudeste"]
df_sudeste = df_sudeste.loc[df['Sexo'] == 'Total']
df_sudeste = df_sudeste.loc[df['Grupo de idade'] == '15 anos ou mais']
df_sudeste = df_sudeste[['Taxa']]
df_sudeste.rename(columns = {'Taxa':'Sudeste'}, inplace = True)
df_sudeste = df_sudeste.reset_index()
df_sudeste = df_sudeste[['Sudeste']]
df_sudeste
```

```python
df_centro = df.loc[df['Brasil e Grande Região'] == "Centro-Oeste"]
df_centro = df_centro.loc[df['Sexo'] == 'Total']
df_centro = df_centro.loc[df['Grupo de idade'] == '15 anos ou mais']
df_centro = df_centro[['Taxa']]
df_centro.rename(columns = {'Taxa':'Centro-Oeste'}, inplace = True)
df_centro = df_centro.reset_index()
df_centro = df_centro[['Centro-Oeste']]
df_centro
```
